**Ejemplo 1**

Resolvamos la primera RR. Hallemos su expresión cerrada o no recurrente.

\begin{equation}
   f_n=f_{n-1}+f_{n-2};\quad f_0 = 3, f_1=6;\quad n\geq2
\end{equation}

 Analizando la RR, nos damos cuenta que es **lineal, homogénea y constante.** Razón por la cual podemos utilizar el método de coeficientes constantes.

\begin{equation}
    f_{n}=\sum_{j=1}^{nr}{\left(\sum_{i=0}^{m_{j}-1}{b_{i+1}n^{i}r_{j}^{n}}\right)}
\end{equation}

Comencemos: Donde encontremos una función predecesora, haremos un cambio de variable, cambiando el subíndice a ser exponente:

\begin{equation}
   f_n=f_{n-1}+f_{n-2}
\end{equation}

\begin{equation}
   x^n=x^{n-1}+x^{n-2}
\end{equation}
Igualamos a 0, y se obtiene

\begin{equation}
   x^n-x^{n-1}-x^{n-2}=0
\end{equation}

Ahora sacamos como factor común el exponente más distante, es decir:

\begin{equation}
   x^{n-2}(x^2-x-1)=0
\end{equation}

Dado que $x^{n-2} \neq 0$. Nos queda entonces que $x^2-x-1=0$. Por lo cual se deben hallar sus raíces, aplicando la fórmula general.

**Nota:** 

\begin{equation}
  x = \frac{-b\pm\sqrt{b^2-4ac}}{2a}
\end{equation}

Una vez aplicada la fórmula, nos quedan las siguientes raíces:

\begin{equation}
  x_1 = \frac{1+\sqrt{5}}{2}, \quad x_2 = \frac{1-\sqrt{5}}{2}
\end{equation}

La multiplicidad de cada raíz es 1. Aplicando el teorema de coeficientes constantes, tenemos que:

\begin{equation}
   f_n=b_1 n^0 \left(\frac{1+\sqrt{5}}{2}\right)^n + b_2 n^0 \left(\frac{1-\sqrt{5}}{2}\right)^n
\end{equation}
Simplificando,

\begin{equation}
   f_n=b_1 \left(\frac{1+\sqrt{5}}{2}\right)^n + b_2 \left(\frac{1-\sqrt{5}}{2}\right)^n
\end{equation}
Utilizando los casos bases, se llega a lo siguient:

Para $f_0 = 3$

\begin{equation}
   f_0=b_1 \left(\frac{1+\sqrt{5}}{2}\right)^0 + b_2 \left(\frac{1-\sqrt{5}}{2}\right)^0 = 3
\end{equation}

\begin{equation}
   b_1 \underbrace{\left(\frac{1+\sqrt{5}}{2}\right)^0}_{1} + b_2  \underbrace{\left(\frac{1-\sqrt{5}}{2}\right)^0}_{1} = 3
\end{equation}

\begin{equation}
  b_1 + b_2 = 3;\implies b_1 = 3 - b_2
\end{equation}

Para $f_1 = 6$

\begin{equation}
   f_1=( 3 - b_2) \left(\frac{1+\sqrt{5}}{2}\right)^1 + b_2 \left(\frac{1-\sqrt{5}}{2}\right)^1 = 6
\end{equation}

Aplicando álgebra básica:

\begin{equation}
   3\left(\frac{1+\sqrt{5}}{2}\right) -b_2\left(\frac{1+\sqrt{5}}{2}\right)+ b_2 \left(\frac{1-\sqrt{5}}{2}\right) = 6
\end{equation}

\begin{equation}
    b_2\left[-\left(\frac{1+\sqrt{5}}{2}\right)+\left(\frac{1-\sqrt{5}}{2}\right)\right] = 6 - 3\left(\frac{1+\sqrt{5}}{2}\right)
\end{equation}

Resolviendo,
\begin{equation}
  b_2=\frac{15-9\sqrt{5}}{10}
\end{equation}
Por lo cual,

\begin{equation}
  b_1=3-b_2 = 3 - \frac{15-9\sqrt{5}}{10} = \frac{15+9\sqrt{5}}{10}
\end{equation}

Una vez encontradas las constantes, se reemplazan en $f_n$, y se ha hallado la versión cerrada de la RR.
\begin{equation}
   f_n=b_1 \left(\frac{1+\sqrt{5}}{2}\right)^n + b_2 \left(\frac{1-\sqrt{5}}{2}\right)^n
\end{equation}
Finalmente, 
\begin{equation}
   \boxed{f_n=\left(\frac{15+9\sqrt{5}}{10}\right) \left(\frac{1+\sqrt{5}}{2}\right)^n + \left(\frac{15-9\sqrt{5}}{10}\right)\left(\frac{1-\sqrt{5}}{2}\right)^n}
\end{equation}

Ahora comparemos la versión no recurrente con su RR:

```
import numpy as np

def recurrencia(n):
  if n == 0: return 3
  elif n == 1: return 6
  return recurrencia(n-1) + recurrencia(n-2)

def no_recurrente(n):
  return int((15+9*np.sqrt(5))/10 * ((1+np.sqrt(5))/2)**n + (15-9*np.sqrt(5))/10 * ((1-np.sqrt(5))/2)**n)

for i in range(0, 9):
  print("n={i}. Resultado recurrente: {rr}\t Resultado no recurrente: {nrr}".format(i=i, rr=recurrencia(i), nrr=no_recurrente(i)))
```

**Ejemplo 2**

Resolvamos la segunda RR:

\begin{equation}
  f_n = 2f_{n-2} + f_{n-1}; \quad f_0=f_1=1; \quad n \geq 2.
\end{equation}

Dada a sus propiedades, lineal, homogénea y constante. Utilizaremos el método de los coeficientes.

Nuevamente, se cambian los subíndices por exponentes:

\begin{equation}
  x^n = 2x^{n-2} + x^{n-1}
\end{equation}

Se saca como factor el exponente más lejano:

\begin{equation}
  x^n - 2x^{n-2} - x^{n-1} = 0
\end{equation}

\begin{equation}
  x^{n-2}(x^2 - x - 2) = 0
\end{equation}

Dado que $x^{n-2} \neq 0 $, entonces $x^2 - x - 2 = 0$. Hallando las raíces:

\begin{equation}
  (x-2)(x+1) = 0;\quad x_1 = 2, \quad x_2 = -1.
\end{equation}

Una vez halladas las raíces, se procede a hallar la ecuación con sus coeficientes.

\begin{equation}
    f_{n}=\sum_{j=1}^{nr}{\left(\sum_{i=0}^{m_{j}-1}{b_{i+1}n^{i}r_{j}^{n}}\right)}
\end{equation}

\begin{equation}
  f_n = b_1 n^0 (2)^n + b_2 n^0 (-1)^n
\end{equation}
Simplificando

\begin{equation}
  f_n = b_1 (2)^n + b_2 (-1)^n
\end{equation}

Se iguala a los casos bases, y se despeja cada coeficiente:

Para $f_0 = 1$

\begin{equation}
  f_0 = b_1 (2)^0 + b_2 (-1)^0 = 1
\end{equation}

\begin{equation}
  f_0 = b_1 + b_2 = 1; \quad b_1 = 1 - b_2
\end{equation}

Para $f_1 = 1$

\begin{equation}
  f_1 = (1 - b_2) (2)^1 + b_2 (-1)^1 = 1
\end{equation}

\begin{equation}
  (1 - b_2) (2) + b_2 (-1) = 1
\end{equation}

\begin{equation}
  2 -2b_2 - b_2 = 1
\end{equation}

\begin{equation}
  3b_2 = 1; \quad b_2 = \frac{1}{3}
\end{equation}

Hallamos $b_1$

\begin{equation}
  b_1 = 1 - b_2 = 1 - \frac{1}{3} = \frac{2}{3}
\end{equation}

Reemplazamos los coeficientes encontrados en la función no recurrente $f_n$ y se simplifica en caso de ser necesario:

\begin{equation}
  f_n = \frac{2}{3} (2)^n + \frac{1}{3} (-1)^n
\end{equation}

\begin{equation}
  \boxed{f_n = \frac{(2)^{n+1} + (-1)^n}{3}}
\end{equation}

Se comprueba que la solución hallada es la correcta, por medio de su comparación con la versión no recurrente. Se recomienda hacer el mismo proceso por medio de inducción matemática.


```
def recurrencia(n):
  if n == 0: return 1
  elif n == 1: return 1
  return 2*recurrencia(n-2) + recurrencia(n-1)

def no_recurrente(n):
  return int((2**(n+1)+(-1)**n)/3)

for i in range(0, 9):
  print("n={i}. Resultado recurrente: {rr}\t Resultado no recurrente: {nrr}".format(i=i, rr=recurrencia(i), nrr=no_recurrente(i)))
```

**Ejemplo 3**

Dada la relación de recurrencia $f_n = \sqrt{\frac{f_{n-2}}{f_{n-1}}};\quad f_0 = 2^3, f_1 = 2^{-3/2};\quad n\geq 2$. Halle su versión no recurrente.

Para resolver esta RR, se utilizará un cambio de variables. Pero primero, se sacará el logaritmo en base 2 a ambas partes de la ecuación.

\begin{equation}
  \log_2{f_n} = \log_2{\left(\frac{f_{n-2}}{f_{n-1}}\right)^\frac{1}{2}}
\end{equation}

\begin{equation}
  \underbrace{\log_2{f_n}}_{g_n} = \frac{1}{2}\log_2{\left(\frac{f_{n-2}}{f_{n-1}}\right)}
\end{equation}

En este caso, se recomienda hacer un cambio de variable de la forma:

\begin{equation}
  g_n = \log_2{f_n}
\end{equation}

Recordemos que $\log_n(\frac{a}{b}) = \log_n{a} - \log_n{b}$. Entonces,

\begin{equation}
  \underbrace{\log_2{f_n}}_{g_n} = \frac{1}{2}\log_2{f_{n-2}} - \frac{1}{2}\log_2{f_{n-1}}
\end{equation}

\begin{equation}
  \underbrace{\log_2{f_n}}_{g_n} = \frac{1}{2}\underbrace{\log_2{f_{n-2}}}_{g_{n-2}} - \underbrace{\frac{1}{2}\log_2{f_{n-1}}}_{g_{n-1}}
\end{equation}

Reemplazando,

\begin{equation}
  g_n = \frac{1}{2}(g_{n-2}- g_{n-1})
\end{equation}

\begin{equation}
  2g_n = g_{n-2}- g_{n-1}
\end{equation}

La función resultante **sí** es lineal, homogénea y constante. Razón por la cual, se puede utilizar el método de los coeficientes. No obstante, hay que calcular los nuevos casos bases.

\begin{equation}
  g_n = \log_2{f_n}
\end{equation}

Entonces 
\begin{equation}
  g_0 = \log_2{f_0} = \log_2{2^3} = 3
\end{equation}

\begin{equation}
  g_1 = \log_2{f_1} = \log_2{2^{-3/2}} = -\frac{3}{2}
\end{equation}

En este sentido, la nueva RR. queda de la siguiente forma: 

\begin{equation}
  2g_n = g_{n-2}- g_{n-1}; \quad g_0 = 3, g_1 = -\frac{3}{2};\quad n\geq2
\end{equation}

Se cambian los suníndices por exponentes:

\begin{equation}
  2x^n = x^{n-2}- x^{n-1}
\end{equation}

\begin{equation}
  x^{n-2}(2x^2 + x - 1) = 0
\end{equation}

De igual forma que los ejercicios anteriores, la expresión es nula cuando $2x^2 + x - 1=0$.

\begin{equation}
  x_1 = \frac{1}{2},\quad x_2 = -1
\end{equation}

Se halla la expresión $g_n$

\begin{equation}
    g_{n}=\sum_{j=1}^{nr}{\left(\sum_{i=0}^{m_{j}-1}{b_{i+1}n^{i}r_{j}^{n}}\right)}
\end{equation}

\begin{equation}
  g_n = b_1 n^0 \left(\frac{1}{2}\right)^n + b_2 n^0 (-1)^n
\end{equation}

\begin{equation}
  g_n = b_1 \left(\frac{1}{2}\right)^n + b_2(-1)^n
\end{equation}

Igualando a los casos bases se obtiene que:

Para $g_0 = 3$

\begin{equation}
  g_0 = b_1 \left(\frac{1}{2}\right)^0 + b_2(-1)^0 = 3
\end{equation}

\begin{equation}
  g_0 = b_1  + b_2 = 3; \quad b_1 = 3 - b_2
\end{equation}

Para $g_1 = -3/2$

\begin{equation}
  g_1 = (3 - b_2) \left(\frac{1}{2}\right)^1 + b_2(-1)^1 = -\frac{3}{2}
\end{equation}

\begin{equation}
  3\left(\frac{1}{2}\right) - b_2\left(\frac{1}{2}\right) - b_2 = -\frac{3}{2}
\end{equation}

\begin{equation}
  3\left(\frac{1}{2}\right) - b_2\left(\frac{3}{2}\right) = -\frac{3}{2}
\end{equation}

\begin{equation}
  3- 3b_2 = -3
\end{equation}

\begin{equation}
  b_2 = 2; \quad b_1 = 1
\end{equation}

Una vez calculados los coeficientes, se reemplzan en la función $g_n$

\begin{equation}
  g_n = \left(\frac{1}{2}\right)^n + 2(-1)^n
\end{equation}

**Nota:** 

 $\log_a{b}=c$ y $a^c = b$

 Recordando que 

\begin{equation}
  g_n = \log_2{f_n}
\end{equation}

Nos queda lo siguiente


\begin{equation}
  \boxed{f_n = 2^{\left[\left(\frac{1}{2}\right)^n + 2(-1)^n\right]}}
\end{equation}

Una vez hallada $f_n$, se comprueba que corresponde a lo esperado.




```
import numpy as np

def recurrencia(n):
  if n == 0: return 2**3
  elif n == 1: return 2**(-3/2)
  return np.sqrt(recurrencia(n-2)/recurrencia(n-1))

def no_recurrente(n):
  return 2**((0.5)**n + 2*(-1)**n)

for i in range(0, 9):
  print("n={i}. Resultado recurrente: {rr}\t Resultado no recurrente: {nrr}".format(i=i, rr=recurrencia(i), nrr=no_recurrente(i)))
```
