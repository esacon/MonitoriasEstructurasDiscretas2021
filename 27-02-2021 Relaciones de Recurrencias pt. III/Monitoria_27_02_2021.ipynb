**Ejemplo 1**

Dada la RR, halle su expresión cerrada.
\begin{equation}
  f_n=\sqrt{1+ f_{n-1}^2};\quad f_0 = 0;\quad n \geq 1
\end{equation}

Dada que la expresión no es lineal ni homogénea, nos queda resolverlo por el método de iteraciones.

El primer paso propuesto para la solución, es elevar ambas partes al cuadrado.

\begin{equation}
  f_n=\sqrt{1+ f_{n-1}^2};\quad f_0 = 0;\quad n \geq 1
\end{equation}

\begin{equation}
  f_n^2=1+ f_{n-1}^2;\quad f_0 = 0;\quad n \geq 1
\end{equation}

Dada que la expresión es cuadrática y no lineal, se recomienda hacer un cambio de variable $g_n = f_n^2$. Y se calculan los nuevos bases:

$g_n = f_n^2$

$g_0 = f_0^2 = 0^2 = 0.$

\begin{equation}
  g_n=1+ g_{n-1};\quad g_0 = 0;\quad n \geq 1
\end{equation}

Una vez se tiene la nueva ecuación, se procede a resolver:

Para k = 1:

\begin{equation}
  g_n= 1+ g_{n-1};\quad g_0 = 0;\quad n \geq 1
\end{equation}

Para k = 2:

\begin{equation}
  g_n=1+ \underbrace{(1+ g_{(n-1)-1})}_{g_{n-1}}
\end{equation}

\begin{equation}
  g_n=1+ (1+ g_{n-2})
\end{equation}

\begin{equation}
  g_n=2 + g_{n-2}
\end{equation}

Para k = 3:
\begin{equation}
  g_n= 1+ \underbrace{(2 + g_{(n-1)-2})}_{g_{n-2}}
\end{equation}

\begin{equation}
  g_n= 3 + g_{n-3}
\end{equation}

Se realizan tantas iteraciones sean necesarias como para encontrar un patrón. 
En este caso, la ecuación parámetrica es la siguiente:

\begin{equation}
  g_n= k + g_{n-k}
\end{equation}

Para resolver, dado que ya está simplificada la expresión, procederemos a igualar los casos bases.

Para $g_{n-k}=g_0=0$. Esto implica que $n-k=0$, por consiguiente $n=k$. Reemplazamos:

\begin{equation}
  g_n= n + 0 = n
\end{equation}

En este sentido $g_n=n$. Dado que el ejercicio pide resolver $f_n$, debemos recordar que el cambio de variable fue $g_n=f_n^2$. Entonces:

\begin{equation}
  f_n = \sqrt{g_n}
\end{equation}

\begin{equation}
  \boxed{f_n = \sqrt{n}}
\end{equation}

Ahora sólo queda comprobar que la solución hallada se cumple para todo $n$. Se deja como labor del estudiante, realizarlo por indcucción matemática.


```python
import numpy as np

def recurrencia(n):
  if n == 0: return 0
  return np.sqrt(1+recurrencia(n-1)**2)

def no_recurrente(n):
  return np.sqrt(n)

for i in range(0, 9):
  print("n={i}. Resultado recurrente: {rr}\tResultado no recurrente: {nrr}".format(i=i, rr=recurrencia(i), nrr=no_recurrente(i)))
```

**Ejemplo 2**

Dada la RR, halle su expresión no recurrente.

\begin{equation}
  f_n = 6f_{n-1} - 9f_{n-2};\quad f_0=0, f_1=1; \quad n \geq 2.
\end{equation}

Puesto que la RR es lineal, homogénea y constante, se resolverá por el método de los coeficientes.

\begin{equation}
    f_{n}=\sum_{j=1}^{nr}{\left(\sum_{i=0}^{m_{j}-1}{b_{i+1}n^{i}r_{j}^{n}}\right)}
\end{equation}

Cambiemos los subíndices por exponentes.

\begin{equation}
  x^n = 6x^{n-1}-9x^{n-2}
\end{equation}

Igualamos a 0.

\begin{equation}
  x^n - 6x^{n-1}+9x^{n-2} = 0
\end{equation}

Se saca como factor común el exponente más lejano.

\begin{equation}
  x^{n-2}(x^2 - 6x + 9) = 0
\end{equation}

Debido a que $x^{n-2} \neq 0$. Nos queda $x^2 - 6x + 9 = 0$. Hallamos las raíces:

\begin{equation}
  (x-3)^2 = 0; \quad x_1=x_2=3
\end{equation}

Dado que las raíces encontradas tienen multiplicidad mayor que 1. La función $f_n$ queda de la siguiente forma:

\begin{equation}
  f_n = b_1 n^0 (3)^n + b_2 n^1 (3)^n
\end{equation}

\begin{equation}
  f_n = b_1 (3)^n + b_2 n(3)^n
\end{equation}

Igualamos a los casos bases.

Para $f_0 = 0$
\begin{equation}
  f_0 = b_1 (3)^0 + b_2 (0)(3)^0 = 0
\end{equation}

\begin{equation}
  b_1  = 0
\end{equation}

Para $f_1 = 1$
\begin{equation}
  f_1 = 3b_2 = 1
\end{equation}

\begin{equation}
  b_2 = \frac{1}{3}
\end{equation}

Una vez calculados los coeficientes, se reemplazan en la función $f_n$


\begin{equation}
  f_n = b_1 (3)^n + b_2 n(3)^n
\end{equation}

\begin{equation}
  f_n = \frac{1}{3}n(3)^n = 3^{-1}n(3)^n = n(3)^{n-1}
\end{equation}

\begin{equation}
  \boxed{f_n = n(3)^{n-1}}
\end{equation}

Ahora sólo queda comprobar que la solución hallada se cumple para todo $n$. Se deja como labor del estudiante, realizarlo por indcucción matemática.

```python
def recurrencia(n):
  if n == 0: return 0
  elif n == 1: return 1
  return 6*recurrencia(n-1)-9*recurrencia(n-2)

def no_recurrente(n):
  return n*(3)**(n-1)

for i in range(0, 9):
  print("n={i}. Resultado recurrente: {rr}\tResultado no recurrente: {nrr}".format(i=i, rr=recurrencia(i), nrr=no_recurrente(i)))
```
